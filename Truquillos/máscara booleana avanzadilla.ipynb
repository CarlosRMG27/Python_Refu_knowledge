# Trucaxo

Hacemos que un dataframe sea igual a df_ct, pero solo en las
partes donde los elementos de su columna COST_CENTER ORIGINAL
estén contenidos en la columna COST_CENTER ORIGINAL del dataframe df_cd
y hacemos lo mismo para el otro

```python
df_ct=df_ct[df_ct['COST_CENTER ORIGINAL'].isin(df_cd['Ce.coste'])]
df_cd=df_cd[df_cd['Ce.coste'].isin(df_ct['COST_CENTER ORIGINAL'])]
```

```python
lista_cd=df_cd['Ce.coste'].unique().tolist()
lista_ct=df_ct['COST_CENTER ORIGINAL'].unique().tolist()
print('Hay {} elementos únicos en el archivo de consolidado'.format(len(lista_cd)))
print('Hay {} elementos únicos en el archivo de CTMO'.format(len(lista_ct)))
```
