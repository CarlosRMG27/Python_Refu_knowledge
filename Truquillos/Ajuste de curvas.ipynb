```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x=np.linspace(-10,10,100)
x2=(2+np.random.random(100)-1)/20
y=np.exp(x)/(np.exp(x)+1)+x2
plt.plot(x,y)
plt.title('Original')
```

```python
#Esta es la función a la que sospechamos que nuestros datos se ajustan
#Los parámetros que le ponemos son nuestra responsabilidad
#Es imperativo incluir la variable independiente
def func(x,a,b):
    return a*np.exp(b*x)/(np.exp(b*x)+1)
```

```python
c,pcov = curve_fit(func,x,y)
```

```python
print(c)
```

```python
y2=func(x,1.1001924,.77931788)
```

```python
plt.plot(x,y,x,y2)
plt.title('Comparación')
plt.legend(['Original','Ajuste'])
```
