```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('Índices de los países.xlsx')
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

# numpy where

```python
np.where(df['Índice original']>70)
#np.where nos regresa los índices 
#que cumplen la condición
```

```python
np.where(df['Índice original']>70,'Alto','Bajo')
#Podemos crear un nuevo array y asignar un valor según 
#se cumpla o no la condición
```

```python
np.where(df['Índice original']>70,'Alto',df['País'])
```

# pandas where

```python
#np.random.seed(seed=42)
df_num=df[['Índice original', 'Índice con proyección',
       'Cambio porcentual', 'Cambio absoluto']]
```

```python
df_num.head()
```

```python
df_num.where(df_num<70,'Alto')#Se reemplazan los valores que sean falsos
```

```python
#Podemos usarlo con columnas
df_num.where(df_num['Cambio porcentual']>10,'Bajo')
#Esto pone el mismo valor ('Bajo') en todas las columnas 
#donde no se cumpla la condición 
```

```python
df_num.where(lambda x: x<75,'Alto')
```

```python
df_num.where(df_num>90,lambda x: x**2)
#Los valores que no cumplen la condición son reemplazados por sus cuadrados
```
