```python
import pandas as pd
import numpy as np
```

```python
from scipy import stats
```

```python
# En estadística, solemos tener dos hipótesis: nuestra explicación 
#y la hipótesis nula, que dice que nuestra explicación no es suficiente
```

```python
df=pd.read_csv('Grades.csv')
df
```

```python
early_finishers=df[pd.to_datetime(df['assignment1_submission'])<'2016']
early_finishers.head()
```

```python
#Para los late finishers queremos todo lo que no está en early finishers
late_finishers=df[~df.index.isin(early_finishers.index)]
late_finishers.head()
```

```python
print(early_finishers['assignment1_grade'].mean())
print(late_finishers['assignment1_grade'].mean())
```

```python
#Formaremos la hipótesis alternativa de que los números arriba dados son distintos
#Hay que definir una threshold que vamos a aceptar
#Solemos llamarla alpha y es 0.05
from scipy.stats import ttest_ind
```

```python
ttest_ind(early_finishers['assignment1_grade'],late_finishers['assignment1_grade'])
```

```python
#No podemos rechazar la hipótesis nula porque nuestro valor p fue superior a alpha.
#Concluimos que las poblaciones son iguales
```

```python
print(ttest_ind(early_finishers['assignment1_grade'],late_finishers['assignment1_grade']))
print(ttest_ind(early_finishers['assignment2_grade'],late_finishers['assignment2_grade']))
print(ttest_ind(early_finishers['assignment3_grade'],late_finishers['assignment3_grade']))
print(ttest_ind(early_finishers['assignment4_grade'],late_finishers['assignment4_grade']))
print(ttest_ind(early_finishers['assignment5_grade'],late_finishers['assignment5_grade']))
```

```python
#En vez de pvalues, podemos usar intervalos de confianza y análisis bayesiano
```

```python
df1=pd.DataFrame([np.random.random(100) for x in range(100)])
df1
```

```python
df2=pd.DataFrame([np.random.random(100) for x in range(100)])
```

```python
def test(alpha=.1):
    num_diff=0
    for col in df1.columns:
        t,p=ttest_ind(df1[col],df2[col])
        if p<=alpha:
            print('Columna {} estadísticamente significativa a alpha {} y valor p {}'.format(col,alpha,p))
            num_diff+=1
```

```python
test()
```
